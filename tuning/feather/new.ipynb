```python
import matplotlib.pyplot as plt
import pandas as pd

#
# Dữ liệu
intervals = ["[0, 1000]", "(1000, 2000]", "(2000, 3000]"]
frequencies = [50, 30, 20]

# Tạo DataFrame
data = pd.DataFrame({"Khoảng giá trị": intervals, "Tần số": frequencies})

# Vẽ biểu đồ cột
plt.figure(figsize=(8, 5))
plt.bar(data["Khoảng giá trị"], data["Tần số"], width=0.6)
plt.xlabel("Khoảng giá trị")
plt.ylabel("Tần số")
plt.title("Biểu đồ tần số theo nhóm")
plt.show()

```
